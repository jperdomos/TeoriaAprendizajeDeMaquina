```python
class Caja:
    def __init__(self, nombre, prior, naranjas, manzanas):
        """
        Inicializa una caja con su nombre, probabilidad previa (prior) y la cantidad de frutas.
        """
        self.nombre = nombre
        self.prior = prior
        self.naranjas = naranjas
        self.manzanas = manzanas

    def total_frutas(self):
        """Retorna el total de frutas en la caja."""
        return self.naranjas + self.manzanas

    def prob_naranja(self):
        """Calcula la probabilidad de extraer una naranja de esta caja."""
        total = self.total_frutas()
        return self.naranjas / total if total > 0 else 0

    def prob_manzana(self):
        """Calcula la probabilidad de extraer una manzana de esta caja."""
        total = self.total_frutas()
        return self.manzanas / total if total > 0 else 0


class BayesCalculator:
    def __init__(self, caja_roja, caja_azul):
        """
        Inicializa la calculadora de Bayes con dos cajas: roja y azul.
        """
        self.caja_roja = caja_roja
        self.caja_azul = caja_azul

    def prob_total(self, fruta):
        """
        Calcula la probabilidad total de obtener la fruta (naranja o manzana)
        considerando ambas cajas.
        """
        if fruta == 'naranja':
            return (self.caja_roja.prob_naranja() * self.caja_roja.prior +
                    self.caja_azul.prob_naranja() * self.caja_azul.prior)
        elif fruta == 'manzana':
            return (self.caja_roja.prob_manzana() * self.caja_roja.prior +
                    self.caja_azul.prob_manzana() * self.caja_azul.prior)
        else:
            raise ValueError("Fruta no reconocida. Usa 'naranja' o 'manzana'.")

    def bayes(self, fruta, caja):
        """
        Aplica el teorema de Bayes para calcular la probabilidad de que la fruta
        provenga de la caja especificada.
        """
        p_total = self.prob_total(fruta)
        if p_total == 0:
            return 0

        if fruta == 'naranja':
            if caja.nombre.lower() == 'roja':
                return (self.caja_roja.prob_naranja() * self.caja_roja.prior) / p_total
            elif caja.nombre.lower() == 'azul':
                return (self.caja_azul.prob_naranja() * self.caja_azul.prior) / p_total
        elif fruta == 'manzana':
            if caja.nombre.lower() == 'roja':
                return (self.caja_roja.prob_manzana() * self.caja_roja.prior) / p_total
            elif caja.nombre.lower() == 'azul':
                return (self.caja_azul.prob_manzana() * self.caja_azul.prior) / p_total
        return None


def solicitar_datos():
    """Solicita al usuario los datos de los priors y la cantidad de frutas en cada caja."""
    print("Introduce los datos para cada caja:")

    # Datos para la caja roja
    red_prior = float(input("Prior para la caja roja (ej. 0.4): "))
    print("\n--- Datos para la caja roja ---")
    red_naranjas = int(input("Número de naranjas en la caja roja: "))
    red_manzanas = int(input("Número de manzanas en la caja roja: "))

    # Datos para la caja azul
    blue_prior = float(input("\nPrior para la caja azul (ej. 0.6): "))
    print("\n--- Datos para la caja azul ---")
    blue_naranjas = int(input("Número de naranjas en la caja azul: "))
    blue_manzanas = int(input("Número de manzanas en la caja azul: "))

    # Crear objetos Caja
    caja_roja = Caja("roja", red_prior, red_naranjas, red_manzanas)
    caja_azul = Caja("azul", blue_prior, blue_naranjas, blue_manzanas)
    return caja_roja, caja_azul


def mostrar_resultados(calculadora):
    """Calcula y muestra todas las probabilidades condicionales y totales."""

    # Probabilidades dentro de cada caja
    print("\n--- Probabilidades condicionales dentro de cada caja ---")
    print(f"Caja roja: P(naranja|roja) = {calculadora.caja_roja.prob_naranja():.4f}, "
          f"P(manzana|roja) = {calculadora.caja_roja.prob_manzana():.4f}")
    print(f"Caja azul: P(naranja|azul) = {calculadora.caja_azul.prob_naranja():.4f}, "
          f"P(manzana|azul) = {calculadora.caja_azul.prob_manzana():.4f}")

    # Probabilidad total de obtener cada fruta
    p_total_naranja = calculadora.prob_total('naranja')
    p_total_manzana = calculadora.prob_total('manzana')
    print("\n--- Probabilidades totales ---")
    print(f"P(naranja) total = {p_total_naranja:.4f}")
    print(f"P(manzana) total = {p_total_manzana:.4f}")

    # Aplicando el teorema de Bayes
    print("\n--- Aplicando el teorema de Bayes ---")
    p_roja_dada_naranja = calculadora.bayes('naranja', calculadora.caja_roja)
    p_azul_dada_naranja = calculadora.bayes('naranja', calculadora.caja_azul)
    p_roja_dada_manzana = calculadora.bayes('manzana', calculadora.caja_roja)
    p_azul_dada_manzana = calculadora.bayes('manzana', calculadora.caja_azul)

    print(f"P(caja roja|naranja) = {p_roja_dada_naranja:.4f}")
    print(f"P(caja azul|naranja) = {p_azul_dada_naranja:.4f}")
    print(f"P(caja roja|manzana) = {p_roja_dada_manzana:.4f}")
    print(f"P(caja azul|manzana) = {p_azul_dada_manzana:.4f}")


def main():
    # Solicitar datos de entrada
    caja_roja, caja_azul = solicitar_datos()

    # Inicializar la calculadora de Bayes
    calculadora = BayesCalculator(caja_roja, caja_azul)

    # Mostrar resultados
    mostrar_resultados(calculadora)


if __name__ == "__main__":
    main()
```

```python
class Caja:
    def __init__(self, nombre, prior, frutas):
        """
        Inicializa una caja con su nombre, probabilidad previa (prior) y un diccionario de frutas.
        El diccionario 'frutas' debe tener como claves el nombre de la fruta y como valor la cantidad.
        """
        self.nombre = nombre
        self.prior = prior
        self.frutas = frutas  # diccionario {fruta: cantidad}

    def total_frutas(self):
        """Retorna el total de frutas en la caja."""
        return sum(self.frutas.values())

    def prob_fruta(self, fruta):
        """
        Calcula la probabilidad de extraer una fruta específica de esta caja.
        Si la fruta no existe en la caja o el total es cero, retorna 0.
        """
        total = self.total_frutas()
        if total == 0:
            return 0
        return self.frutas.get(fruta, 0) / total


class BayesCalculator:
    def __init__(self, cajas, lista_frutas):
        """
        Inicializa la calculadora con una lista de cajas y una lista de tipos de frutas.
        """
        self.cajas = cajas
        self.lista_frutas = lista_frutas

    def prob_total_fruta(self, fruta):
        """
        Calcula la probabilidad total de obtener la fruta 'fruta' considerando todas las cajas.
        """
        p_total = 0
        for caja in self.cajas:
            p_total += caja.prob_fruta(fruta) * caja.prior
        return p_total

    def bayes(self, fruta, caja):
        """
        Aplica el teorema de Bayes para calcular la probabilidad de que habiendo obtenido
        la fruta 'fruta' ésta provenga de la 'caja' especificada.
        """
        p_total = self.prob_total_fruta(fruta)
        if p_total == 0:
            return 0
        return (caja.prob_fruta(fruta) * caja.prior) / p_total


def solicitar_frutas():
    """
    Solicita al usuario la cantidad de tipos de frutas y sus nombres.
    Retorna una lista con los nombres de las frutas.
    """
    num_frutas = int(input("Ingresa el número de tipos de frutas: "))
    lista = []
    for i in range(num_frutas):
        fruta = input(f"Ingresa el nombre de la fruta {i+1}: ").strip().lower()
        lista.append(fruta)
    return lista


def solicitar_cajas(lista_frutas):
    """
    Solicita al usuario la cantidad de cajas y los datos de cada una.
    Para cada caja se pide:
     - Nombre
     - Prior (por ejemplo, 0.4)
     - Cantidad de cada fruta (según 'lista_frutas')
    Retorna una lista de objetos Caja.
    """
    num_cajas = int(input("Ingresa el número de cajas: "))
    cajas = []
    for i in range(num_cajas):
        print(f"\n--- Datos para la caja {i+1} ---")
        nombre = input("Nombre de la caja: ").strip().lower()
        prior = float(input("Prior para esta caja (ej. 0.4): "))
        frutas = {}
        for fruta in lista_frutas:
            cantidad = int(input(f"Número de '{fruta}' en la caja {nombre}: "))
            frutas[fruta] = cantidad
        cajas.append(Caja(nombre, prior, frutas))
    return cajas


def mostrar_resultados(calculadora):
    """
    Muestra las probabilidades condicionales dentro de cada caja, la probabilidad total
    de obtener cada fruta y las probabilidades aplicando el teorema de Bayes.
    """
    print("\n--- Probabilidades condicionales en cada caja ---")
    for caja in calculadora.cajas:
        print(f"\nCaja '{caja.nombre}':")
        for fruta in calculadora.lista_frutas:
            p = caja.prob_fruta(fruta)
            print(f"  P({fruta}|{caja.nombre}) = {p:.4f}")

    print("\n--- Probabilidades totales de cada fruta ---")
    for fruta in calculadora.lista_frutas:
        p_total = calculadora.prob_total_fruta(fruta)
        print(f"P({fruta}) total = {p_total:.4f}")

    print("\n--- Aplicando el teorema de Bayes ---")
    for fruta in calculadora.lista_frutas:
        print(f"\nPara la fruta '{fruta}':")
        for caja in calculadora.cajas:
            p_bayes = calculadora.bayes(fruta, caja)
            print(f"  P(caja '{caja.nombre}'|{fruta}) = {p_bayes:.4f}")


def main():
    print("=== Cálculo de probabilidades usando el teorema de Bayes ===")
    lista_frutas = solicitar_frutas()
    cajas = solicitar_cajas(lista_frutas)
    calculadora = BayesCalculator(cajas, lista_frutas)
    mostrar_resultados(calculadora)


if __name__ == "__main__":
    main()
```
